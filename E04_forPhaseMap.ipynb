```python
import numpy as np

def create_phase_array(n_pxl, tht, rate):
    ''' 
    Args:
        n_pxl (int): Number of pixels
        tht (float): Angle in radians
        rate (float): Rate
    Returns:
        arr (arr): [2*n_pxl+1, 2*n_pxl+1] arr[n_pxl, n_pxl] = 0 where (n_pxl, n_pxl) is the center.
            in tht direction linear (i,j)->R, the steepness of gradint is rate
    '''
    # 1. Create a coordinate grid from -n_pxl to n_pxl
    # The total size will be 2 * n_pxl + 1
    x = np.arange(-n_pxl, n_pxl + 1)
    y = np.arange(-n_pxl, n_pxl + 1)
    
    # 2. Generate 2D coordinate matrices
    # X and Y represent the distance from the center (n_pxl, n_pxl)
    X, Y = np.meshgrid(x, y)
    
    # 3. Project the coordinates onto the line in the 'tht' direction
    # Using the dot product formula: X * cos(tht) + Y * sin(tht)
    # This aligns the gradient toward the specified angle
    projected_distance = X * np.cos(tht) + Y * np.sin(tht)
    
    # 4. Multiply by the rate to set the steepness of the gradient
    arr = rate * projected_distance
    
    return arr
```
